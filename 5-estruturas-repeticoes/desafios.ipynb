## 2. Crie um programa que conte quantos números pares que estão no intervalo entre 1 e 200 e armazene-os em uma lista.​

```python
intervalo = range(1, 201)
pares = []
for numero in intervalo:
    if numero % 2 == 0:
       # print(f'{numero} é par! Armazenando na lista.')
        pares.append(numero)
print(f'Foram encontrados {len(pares)} números pares entre 1 e 200.')
```

## 3. Faça um programa que calcule a soma entre todos os números que são múltiplos de três e que se encontram no intervalo de 1 até 500.​

```python
intervalo = range(1, 500)
soma = 0
for numero in intervalo:
    if numero % 3 == 0:
        soma += numero
       # print(f'{numero} é múltiplo de 3. Somando...')
print(f'A soma dos números múltiplos de 3 entre 1 e 500 é {soma}.')
```

## 4. Mostre a tabuada de um número digitado pelo usuário.​

## 5. Crie um programa que leia uma frase qualquer e diga se ela é um palíndromo, desconsiderando os espaços.
Palíndromos são frases ou palavras que podem ser lidas, indiferentemente, da esquerda para a direita, conforme o sentido habitual da leitura, ou da direita para a esquerda, sem que haja mudança da sua significação. São também chamados de anacíclicos.

```python
frase = input('Digite uma frase qualquer: ').replace(' ', '').lower()
inverso = frase[::-1] # Aqui estamos invertendo a string usando slicing
if frase == inverso:
    print('A frase digitada é um palíndromo!')
    print(f'Frase original: {frase}')
    print(f'Frase invertida: {inverso}')
else:
    print('A frase digitada não é um palíndromo.')
    print(f'Frase original: {frase}')
    print(f'Frase invertida: {inverso}')
```

## 6. Crie um programa que leia o ano de nascimento de sete pessoas. No final, mostre quantas pessoas ainda não atingiram a maioridade e quantas já são maiores.​

```python
dict_pessoas = {'Lucas': 1989, 'Ana': 2005, 'Pedro': 2010, 'Mariana': 1995, 'João': 2000, 'Carla': 2015, 'Rafael': 1985}
maiores = 0
menores = 0
ano_atual = 2025
for nome, ano_nascimento in dict_pessoas.items(): # Notem que aqui estamos usando o método items() para obter tanto a chave quanto o valor
    idade = ano_atual - ano_nascimento
    if idade >= 18:
        maiores += 1
        print(f'{nome} tem {idade} anos e é maior de idade.')
    else:
        menores += 1
        print(f'{nome} tem {idade} anos e é menor de idade.')
print(f'No total, temos {maiores} pessoas maiores de idade e {menores} pessoas menores de idade.')
```

# 7. Faça um programa que leia o peso de cinco pessoas. No final, mostre qual foi o maior e o menor peso lidos.​

```python
dict_pessoas = {'Pedro': 90, 'Ana': 85, 'João': 78, 'Maria': 92, 'Lucas': 120}
for nome, peso in dict_pessoas.items():
    print(f'{nome} pesa {peso} kg.')
maior_peso = max(dict_pessoas.values())
menor_peso = min(dict_pessoas.values())
print(f'O maior peso registrado é {maior_peso} kg.')
```

# 9. Crie um programa que leia dois valores e mostre um menu na tela:

[ 2 ] multiplicar

[ 3 ] maior

[ 5 ] sair do programa

Seu programa deverá realizar a operação solicitada em cada caso.

# 10. Em uma competição de ginástica, cada atleta recebe votos de sete jurados. A melhor e a pior nota são eliminadas. A sua nota fica sendo a média dos votos restantes. Você deve fazer um programa que receba o nome do ginasta e as notas dos sete jurados alcançadas pelo atleta em sua apresentação e depois informe a sua média, conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média com as notas restantes). As notas não são informados ordenadas. Um exemplo de saída do programa deve ser conforme o exemplo abaixo:

Atleta: Aparecido Parente
Nota: 9.9
Nota: 7.5
Nota: 9.5
Nota: 8.5
Nota: 9.0
Nota: 8.5
Nota: 9.7
Resultado final:
Atleta: Aparecido Parente
Melhor nota: 9.9
Pior nota: 7.5
Média: 9,04

```python
# 1: Vamos capiturar o nome do atleta
nome_atleta = input('Digite o nome do atleta: ')
print(f'Atleta registrado: {nome_atleta}')
# 2: Agora vamos capiturar as 7 notas
notas = [] # Criamos uma lista vazia para armazenar as notas
for i in range(7): # Usamos um range de 7 para capturar as 7 notas
    nota = float(input(f'Digite a nota {i + 1} do atleta {nome_atleta}: ')) # Usamos i + 1 para exibir a numeração correta das notas. Ou seja o iniciará em 0, mas exibiremos para o usuário começando em 1
    notas.append(nota) # Adicionamos a nota à lista
    print(f'Nota {i + 1} registrada: {nota}')
# 3: Agora vamos calcular a média, descartando a maior e a menor nota
maior_nota = max(notas)
menor_nota = min(notas)
notas.remove(maior_nota)
notas.remove(menor_nota)
media = sum(notas) / len(notas) # Calculamos a média das notas restantes
# 4: Finalmente, vamos exibir o resultado
print(f'O atleta {nome_atleta} teve as seguintes notas: {notas}')
print(f'A maior nota descartada foi: {maior_nota}')
print(f'A menor nota descartada foi: {menor_nota}')
print(f'A média das notas restantes é: {media:.2f}')
```

```python
dicionario = {1: 'a', 2: 'b', 'abc': 100, False: (2 + 3j)}
dicionario
```

```python
for elem in dicionario:
    print(elem)
```

```python
for elem in dicionario.values():
    print(elem)
```

```python
for elem in dicionario.items():
    print(elem)
```

```python
for key, value in dicionario.items():
    print(f'Chave: {key} | Valor: {value}')
```

```python
dicionario2 = {'a': 1, 'b': 2, 'c': 3}
for x, y, z in dicionario2.items():
    print(x, y)
```

O erro ocorre, pois o comando for x, y, z in dicionario.items() tenta desempacotar três variáveis, mas um dicionário só fornece duas informações em cada iteração: a chave e o valor. Portanto, isso resulta em um erro na execução do código.

```python
a = 0

while a < 10:
    print(a)
    if a == 5:
        break
    a += 2
    print('_'*10)
```

```python
a = 0

while a < 10:
    if a % 2 == 1:
        a += 1
        continue

    print(a)
    a += 1
```

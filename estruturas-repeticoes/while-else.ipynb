# Estrutura de Repetições While
O while é uma estrutura de repetição em várias linguagens de programação que permite executar um bloco de código enquanto uma condição específica for verdadeira. A sintaxe básica do while é a seguinte:

```python
while condição:
    # bloco de código a ser executado
```
É muito importante garantir que a condição eventualmente se torne falsa, caso contrário, o loop continuará indefinidamente, resultando em um loop infinito.

### Exemplo de Uso do While

````python
f = 'curso de python'
g = len(f)
g
````

Agora vamos fazer um slice da string do começo até o comprimento g e vamos verificar se a string começa com 'curso de '. E para que a condição se torne falsa vamos decrementar o valor de g a cada iteração do loop.

````python
while f[:g].startswith('curso de '):
    print(f[:g])
    g -= 1
else:
    print(f[:g] + 'não começa mais com "curso de "')
````

Na aula anterior, aprendemos como definir uma função e sua estrutura. Agora vamos explorar como retornar valores a partir de uma função usando a palavra-chave `return`.

```python
def soma(x,y):
    return x + y
```

```python
soma(10, 5)
```

Se não usarmos `return`, a função retornará `None` por padrão. Por isso é necessário utilizar `return` quando queremos que a função nos dê um resultado específico.

```python
def ola(nome='Lucas'):
    return f'Olá {nome}'
```

```python
lucas = ola()
```

```python
lucas
```

```python
jose = ola('José')
```

```python
jose
```

Note que após o `return`, a função encerra sua execução, ou seja, qualquer código após o `return` não será executado. E que com o return podemos retornar qualquer tipo de dado: números, strings, listas, dicionários, etc e salvamos esse valor em uma variável para uso posterior. Sem o `return`, a função apenas executa o código dentro dela, mas não nos dá nenhum valor de volta e com isso não conseguimos salvar o resultado em uma variável.

Agora vamos criar uma função que vai receber o nome de um aluno e as notas que ele tirou e vai retornar a média dessas notas.

```python
def media_notas_aluno(nome, notas):
    media = sum(notas) / len(notas)
    return f'A média do aluno {nome} é {media:.2f}'
```

Agora vamos estudar o return com valores multiplos.

```python
def soma(x, y):
    print('Comecei a função')
    return x + y
    print('Terminei a função')
```

```python
soma(2, 3)
```

Note que após o `return`, a função encerra sua execução, ou seja, qualquer código após o `return` não será executado.

```python
def calculadora(x, y, operacao):
    if operacao == 'soma':
        return x + y
    elif operacao == 'subtracao':
        return x - y
    elif operacao == 'multiplicacao':
        return x * y
    elif operacao == 'divisao':
        return x / y
    else:
        return 'Operação inválida'
```

Nós usamos o if pois se a gente tivesse colocado um return após o outro para as operações, a função iria encerrar na primeira que fosse verdadeira, sem verificar as outras condições porque é assim que o return funciona. O que poderia ter sido feito é colocar após o `return` as operações separadas por vírgula, assim a função retornaria uma tupla com os três resultados. Mas nesse caso, queremos retornar apenas o resultado da operação escolhida pelo usuário.

```python
calculadora(10, 5, 'soma')
```

```python
def foo():
    return [1,2,3], 'Lucas', 3.14, 10, True
```

```python
lista, string, flutuante, inteiro, booleano = foo()
```

```python
lista
```

```python
booleano
```

No exemplo acima, mostramos como o python permite retornar múltiplos valores de uma função, retornando-os em uma tupla implícita. Podemos desempacotar esses valores em variáveis separadas ao chamar a função, facilitando o acesso a cada valor individualmente.

# Variáveis Globáis e Locais

Quando usamos funções, precisamos nos atentar às variáveis. Variáveis locais existem somente dentro da função onde foram instanciadas, já variáveis globais existem fora da função e podem ser acessadas por todas as funções e módulos. Observe:

```python
x = 10  # Variável global
def minha_funcao():
    y = 5  # Variável local
    print("Dentro da função, y =", y)
    print("Dentro da função, x =", x) # Acessando variável global
minha_funcao()
print("Fora da função, x =", x)
# print("Fora da função, y =", y)  # Isso causará um erro
```
Neste exemplo, `x` é uma variável global e pode ser acessada dentro da função `minha_funcao`. Já `y` é uma variável local e só pode ser acessada dentro dessa função. Tentar acessar `y` fora da função resultará em um erro, pois ela não existe nesse escopo.

````python
comida = "arroz"

def muda_comida():
    global comida # avisando que queremos a variável global
    comida = "batata"
    print("Comida: {nova_comida}".format(nova_comida=comida))

print(comida) # comida antes de mudar
# 'arroz'

muda_comida() # chamando a função
# 'Comida: batata'

print(comida) # comida depos da função muda_comida()
# 'batata'
````

No exemplo acima temos uma variável global chamada comida. Dentro da função muda_comida, usamos a palavra-chave global para indicar que queremos modificar a variável global comida, e não criar uma nova variável local com o mesmo nome. Após chamar a função, a variável global comida é atualizada para "batata".

# Definição de uma Função

![Estrutura de uma função](/home/lucasfaust/estudos/python-course-udemy/6-funcoes/image.png)

A estrutura de uma função em Python é composta por:
- A palavra-chave `def`, que indica o início da definição da função.
- O nome da função, que deve seguir as regras de nomenclatura de variáveis.
- Parênteses `()`, que podem conter parâmetros separados por vírgulas. Aqui os parênteses estão vazios, indicando que a função não recebe parâmetros. Eles servem para passar informações para a função.
- Dois pontos `:`, que indicam o início do bloco de código da função.
- O bloco de código indentado, que contém as instruções que serão executadas quando a função for chamada. A identação tem a finalidade de definir o escopo da função.
- O Assigmentment statement value = 1, indica que a variável `value` está sendo atribuída com o valor `1`. Dentro da função, você pode ter várias linhas de código que realizam diferentes tarefas.
- O return statement `return value`, que especifica o valor que a função irá retornar quando for chamada. Neste caso, a função retorna o valor da variável `value`.

OBS: A função não é executada quando é definida. Ela só será executada quando for chamada em outro ponto do código, utilizando o nome da função seguido de parênteses, como `foo()`.


```python
def foo():
    valor = 1
    print(valor)
```

```python
foo()
```

Notem que a função cria a variável `value` dentro do seu escopo local. Essa variável não é acessível fora da função, a menos que seja retornada ou declarada como global (o que não é o caso aqui).

```python
def anfitriao(nome):
    print(f'Olá, {nome}! Seja bem-vindo(a)!')
```

```python
anfitriao('Lucas')
```

As funções podem receber parâmetros. Eles são valores que você pode passar para a função quando a chama, permitindo que a função trabalhe com diferentes dados. No exemplo acima a função anfitriao recebe um parâmetro chamado `nome`. Quando você chama a função, você pode fornecer um valor para esse parâmetro, como mostrado na chamada `anfitriao('Lucas')`. Isso faz com que a função imprima uma mensagem personalizada de boas-vindas usando o nome fornecido.

```python
def anfitriao2(nome, idade, sexo):
    print(f'Olá, {nome}! Você tem {idade} anos e seu sexo é {sexo}.')
```

```python
anfitriao2('Lucas', 37, 'Masculino')
```

```python
anfitriao2(idade=28, sexo='Feminino', nome='Ana')
```

Note que na chamada da função acima, mudamos a ordem dos parâmetros ao usar nomes explícitos (`idade=28`, `sexo='Feminino'`, `nome='Ana'`). Isso é útil quando você tem muitos parâmetros ou quando deseja tornar o código mais legível, permitindo que você especifique os valores dos parâmetros em qualquer ordem.

```python
def em_cartas(filmes=['Stars Wars', 'Matrix', 'O Senhor dos Anéis']):
        print(f'Os filmes em cartaz são: {filmes}')
```

```python
em_cartas()
```

Podemos notar que ao definir a função `em_cartas`, foi atribuído um valor padrão ao parâmetro `filmes`. Isso significa que, se a função for chamada sem fornecer um argumento para `filmes`, ela usará o valor padrão definido na definição da função. No exemplo acima, quando chamamos `em_cartas()` sem argumentos, a função imprime a lista padrão de filmes: `['Stars Wars', 'Matrix', 'O Senhor dos Anéis']`. Se quisermos passar uma lista diferente de filmes, podemos fazer isso ao chamar a função com um argumento, como `em_cartas(['Inception', 'Avatar'])`, e a função então usará essa nova lista em vez do valor padrão.

```python
em_cartas(['Inception', 'Avatar'])
```
